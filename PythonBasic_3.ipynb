```python
import numpy as np
import pandas as pd
```

```python
mtcars = pd.read_csv("mtcars.csv")
print(mtcars.head())
mtcars.index = mtcars.model
del mtcars["model"]
mtcars
```

```python
book = pd.read_csv("draft-saved.csv")
del book["Player"]  #del để xóa
book
```

```python
#mean : tính trung bình cộng
sum_car=mtcars.sum()
shape_car=mtcars.shape[0]
colmeans=sum_car/shape_car
colmeans
```

```python
centered = mtcars - colmeans
centered.describe()
```

```python
from sklearn import preprocessing
# chuẩn hóa tập dữ liệu dọc theo bất kì trục nào,chuẩn hóa các điểm dữ liệu. 
# Nó sẽ chia cho độ lệch chuẩn và trừ đi giá trị trung bình cho mỗi điểm dữ liệu.
scale_data = preprocessing.scale(mtcars)
scale_cars = pd.DataFrame(scale_data,index=mtcars.index,columns=mtcars.columns)
scale_cars.describe()
```

```python
normally = np.random.normal(size=1000) #vẽ mẫu ngẫu nhiên từ phân phối chuẩn(size)
normally = pd.DataFrame(normally)
normally.hist(figsize=(8,8),bins=30) 
#hist là biểu đồ phân bố tần suất
```

```python
#Exponential Distribution
from numpy import random

x = random.exponential(scale=5, size=10000)
#exponential: phân phối hàm mũ
#scale: tỉ lệ
#size: số phần tử, ví dụ, size=3 thì khi print sẽ có 3 phần tử

# print(x)
x = pd.DataFrame(x)
x.hist(figsize=(8,8),bins=10)
```
